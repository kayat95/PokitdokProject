```python
##SOURCE https://www.cms.gov/Research-Statistics-Data-and-Systems/Statistics-Trends-and-Reports/Medicare-Provider-Charge-Data/index.html

medicarePUF2 = pd.read_csv('./PhysicianCompare2014.csv')
```

```python
cols = medicarePUF2.columns.tolist()[4:]
cols.append(medicarePUF2.columns.tolist()[1])
```

```python
def PUFtoJSON(j):
    PUF = {}
    for col in cols:
        if(not(np.isnan(medicarePUF2[col][j]))):
            PUF[col] = str(medicarePUF2[col][j])
        
    return PUF
    print(PUF)
```

```python
#gets unique NPIs from PUF
NPIs = set(medicarePUF2.NPI)
```

```python
update = {}
#initialize update
def createRecordDict():
    for npi in NPIs:
        update[str(npi)] = list()
    return update
   # print(update)
```
